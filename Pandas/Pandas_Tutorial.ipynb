## 08. Merge Dataframes

#### Basic Merge using a Dataframe Column

```python
import pandas as pd
df_1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35]
})
df_1
```

```python
df_2 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "humidity": [65, 68, 75]
})
df_2
```

```python
df_3 = pd.merge(df_1, df_2, on = "city")
df_3
```

#### Types of Database Joins

```python
df_1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 32]
})
print(df_1, "\n")

df_2 = pd.DataFrame({
    "city": ["new york", "chicago", "san francisco"],
    "humidity": [65, 68, 75]
})
print(df_2, "\n")
```

```python
# Note that only common part is there i.e. the intersection part of two dataframes
# In database concept it is an inner join
df_3 = pd.merge(df_1, df_2, on = "city")
df_3
```

```python
# Note that total part is there i.e. the union part of two dataframes
# In database concept it is an outer join, by mentioning (how = "outer")
df_3 = pd.merge(df_1, df_2, on = "city", how = "outer")
df_3
```

```python
# Note that the df_1 part is there i.e. the first part of two dataframes
# In database concept it is an left join, by mentioning (how = "left")
df_3 = pd.merge(df_1, df_2, on = "city", how = "left")
df_3
```

```python
# Note that the df_2 part is there i.e. the second part of two dataframes
# In database concept it is an right join, by mentioning (how = "right")
df_3 = pd.merge(df_1, df_2, on = "city", how = "right")
df_3
```

#### Indicator Flag

```python
# Indicator of the dataframes from which the cities are comming
df_3 = pd.merge(df_1, df_2, on = "city", how = "outer", indicator = True)
df_3 # look at the "_merge" column
```

#### Suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 38],
    "humidity": [65, 68, 71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "san diego"],
    "temperature": [21, 14, 35],
    "humidity": [65, 68, 71]
})
df2
```

```python
df3= pd.merge(df1, df2, on = "city", how = "outer")
df3 # Column values are repeating, that's why there are indexing in merged dataframe
```

```python
df3= pd.merge(df1, df2, on = "city", how = "outer", suffixes = ('_first', '_second'))
df3 # User defined suffixes for column names in merged dataframe
```

#### Join

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
print(df1, "\n")

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
print(df2, "\n")

df1.join(df2,lsuffix='_l', rsuffix='_r')
```
